```
# meng-import library networkx dan matplotlib
import networkx as nx
import matplotlib.pyplot as plt

# defini fungsi shortest_path yang nantinya akan diberikan value berupa graph, source dan dest
# graph merupakan grafik yang berisikan node-node
# source merupakan titik/node awal/pertama
# dest merupakan titik/node terakhir
def shortest_path(graph, source, dest):
    # mendefisinikan variabel array result
    result = []
    # variabel result akan langsung diisi dengan source
    result.append(source)
    # membuat variabel jumlah untuk menampung jumlah jarak yang ditempuh
    jumlah = 0
    # ketika dest belum ada di array result, maka perulangan akan terus dilakukan
    while dest not in result:
        # current_node berisikan nilai ke-(-1) dari variabel result, atau nilai terakhir
        current_node = result[-1]

        # local_max merupakan variabel   isi dari node yang tengah di eksekusi
        local_max = min(graph[current_node].values())
        # terus melakukan kunjungan node dari node saat ini
        for node, weight in graph[current_node].items():
            # jika weight bernilai sama dengan local_max, maka
            if weight == local_max:
                # tulisan dibawah akan dicetak
                print ("Node Selanjutnya :",node,"| Jarak :", weight)
                #setiap terjadi perulangan nilai variabel jumlah akan ditambah dengan nilai variabel weight
                jumlah = jumlah + weight
                # nilai node akan ditambahkan dalam array result
                result.append(node)
    # ketika perulangan berakhir, maka tulisan dibawah akan dicetak
    print ("Jarak Tempuh :", jumlah,"KM ke UPJ")
    
    # nilai result akan dikembalikan
    return result
```

```
# menentukan node dari graph masing-masing rumah
rumah_anayah = {'C': {'A': 2.6, 'E': 2.8},
       'A': {'D': 5},
       'D': {'B': 0.7, 'F': 3.3},
       'E': {'F': 4.4},
       'F': {'H': 2.7},
       'B': {'H': 2.5},
       'H': {'G': 2.4}}

rumah_ari = {'A': {'B': 1.4, 'F': 8.5},
       'B': {'C': 2.9, 'D': 2.1},
       'D': {'E': 2.3},
       'C': {'E': 1.4},
       'E': {'G': 1.3},
       'F': {'G': 7.3}}

rumah_banez = {'K': {'L': 0.7},
       'L': {'M': 1.4, 'P': 2.5},
       'M': {'N': 2.6},
       'P': {'Q': 1.4},
       'Q': {'N': 1.5},
       'N': {'O': 1.9}}

rumah_bagoes = {'A': {'B': 3.7},
       'B': {'C': 3.5, 'J': 4.7},
       'J': {'G': 3.2},
       'E': {'I': 1.4},
       'G': {'E': 2.6},
       'D': {'E': 1.6},
       'F': {'E': 1.2},
       'C': {'D': 1.4, 'F': 1.8}}

rumah_triana = {'R': {'S': 1.9, 'V' : 2.1, 'W' : 2.6},
       'S': {'T' : 1.9},
       'T': {'U': 2.3},
       'V': {'U': 3.1},
       'W': {'U': 2.35},
       'U': {'X': 2.1}}
```

```
# definisi fungsi draw yang nantinya akan diberikan value berupa maps
# variabel maps berisikan nilai pilihan maps user
def draw(maps):
    # variabel g akan berisikan fungsi dari library networkx alias nx.DiGraph
    g = nx.DiGraph()
    # mendefiniskan variabel array color
    color = []
    # mendefiniskan variabel array edgeColor
    edgeColor = []
    # jika maps yang dipilih adalah 1, maka maps akan digambar
    if(maps == int (1)):
      g.add_edge("C", "A", weight="2,6 km")
      g.add_edge("C", "E", weight="2,8 km")
      g.add_edge("A", "D", weight="5 km")
      g.add_edge("D", "B", weight="0,7 km")
      g.add_edge("D", "F", weight="3,3 km")
      g.add_edge("E", "F", weight="4,4 km")
      g.add_edge("F", "H", weight="2,7 km")
      g.add_edge("B", "H", weight="2,5 km")
      g.add_edge("H", "G", weight="2,5 km")
      # pemberian warna pada node, 'g'=green, 'b'=blue, 'r'=red
      color = ['g','b','b','b','b','b','b','r']
      # pemberian warna pada edge/garis, 'g'=green, 'b'=blue, 'r'=red
      edgeColor =['r','k','r','k','r','r','k','r']
    # jika maps yang dipilih adalah 2, maka maps akan digambar
    elif(maps == int (2)):
      g.add_edge("A", "B", weight="1,4 km")
      g.add_edge("A", "F", weight="8,5 km")
      g.add_edge("B", "D", weight="2,1 km")
      g.add_edge("B", "C", weight="2,9 km")
      g.add_edge("D", "E", weight="2,3 km")
      g.add_edge("C", "E", weight="1,4 km")
      g.add_edge("E", "G", weight="1,3 km")
      g.add_edge("F", "G", weight="7,3 km")
      # pemberian warna pada node, 'g'=green, 'b'=blue, 'r'=red
      color = ['g','b','b','b','b','b','r']
      # pemberian warna pada edge/garis, 'g'=green, 'b'=blue, 'r'=red
      edgeColor = ['r','k','r','k','k','r','k']
    # jika maps yang dipilih adalah 3, maka maps akan digambar
    elif(maps == int (3)):
      g.add_edge("k", "L", weight="0,7 km")
      g.add_edge("L", "M", weight="1,4 km")
      g.add_edge("L", "P", weight="2,5 km")
      g.add_edge("P", "Q", weight="1,4 km")
      g.add_edge("Q", "N", weight="1,5 km")
      g.add_edge("M", "N", weight="2,6 km")
      g.add_edge("N", "O", weight="1,9 km")
      # pemberian warna pada node, 'g'=green, 'b'=blue, 'r'=red
      color = ['g','b','b','b','b','b','r']
      # pemberian warna pada edge/garis, 'g'=green, 'b'=blue, 'r'=red
      edgeColor = ['r','r','k','r','k','k','r']
    # jika maps yang dipilih adalah 4, maka maps akan digambar
    elif(maps == int (4)):
      g.add_edge("A", "B", weight="3,7 km")
      g.add_edge("B", "C", weight="3,5 km")
      g.add_edge("C", "D", weight="1,4 km")
      g.add_edge("C", "F", weight="1,8 km")
      g.add_edge("J", "G", weight="3,2 km")
      g.add_edge("B", "J", weight="      4,7 km")
      g.add_edge("G", "E", weight="2,6 km")
      g.add_edge("F", "E", weight="1,2 km")
      g.add_edge("D", "E", weight="1,6 km")
      g.add_edge("E", "I", weight="1,4 km")
      # pemberian warna pada node, 'g'=green, 'b'=blue, 'r'=red
      color = ['g','b','b','b','b','b','b','b','r']
      # pemberian warna pada edge/garis, 'g'=green, 'b'=blue, 'r'=red
      edgeColor = ['r','r','k','r','k','r','k','k','k']
    # jika maps yang dipilih adalah 5, maka maps akan digambar
    elif(maps == int (5)):
      g.add_edge("R", "S", weight="1,9 km")
      g.add_edge("S", "T", weight="1,9 km")
      g.add_edge("R", "V", weight="2,1 km")
      g.add_edge("R", "W", weight="2,6 km")
      g.add_edge("T", "U", weight="2,3 km")
      g.add_edge("V", "U", weight="3,1 km")
      g.add_edge("W", "U", weight="2,35 km")
      g.add_edge("U", "X", weight="2,1 km")
      # pemberian warna pada node, 'g'=green, 'b'=blue, 'r'=red
      color = ['g','b','b','b','b','b','r']
      # pemberian warna pada edge/garis, 'g'=green, 'b'=blue, 'r'=red
      edgeColor = ['r','k','k','r','r','k','k']
    # jika user memilih angka selain 1-5, maka tulisan dibawah akan dicetak
    else :
      print ("Pilihan tidak tersedia")
   
    # variabel pos akan berisikan fungsi yang dipanggil dari nx.shell_layout dan diberikan variabel g didalamnya
    pos = nx.shell_layout(g)
    # membuat garis pada grafik
    edge_labels = { (u,v): d['weight'] for u,v,d in g.edges(data=True) }
    # menggunakan fungsi nx.draw_networks_node dan diberikan variabel g,pos,node_size,node_color
    nx.draw_networkx_nodes(g,pos,node_size=1000, node_color=color)
    # menggunakan fungsi nx.draw_networks_edges dan diberikan variabel g,pos,edge_color
    nx.draw_networkx_edges(g,pos, edge_color=edgeColor)
    # menggunakan fungsi nx.draw_networks_labels dan diberikan variabel g,pos
    nx.draw_networkx_labels(g,pos)
    # menggunakan fungsi nx.draw_networks_edge dan diberikan variabel g,pos,edge_labels
    nx.draw_networkx_edge_labels(g,pos,edge_labels=edge_labels)
    # akan mencetak judul 'MAPS'
    plt.title("MAPS")
    plt.axis("off")
    # akan menampilkan graph
    plt.show()

# definisi fungsi _main   
def _main():
  
    # mencetak tulisan dibawah ini
    print ("Rumah Anggota Kelompok 1 :")
    # menampilkan pilihan serta menampung nilai yang diberikan user
    choose = int(input("Pilih Rumah (1-5) :\n1. Rumah Anayah\n2. Rumah Ari\n3. Rumah Banez\n4. Rumah Bagoes\n5. Rumah Triana\n\n"))
    # jika user memilih 1
    if (choose == 1):
        # tulisan dibawah akan dicetak
        print ("\nRumah Anayah ke UPJ:")
        # memanggil dan memberikan nilai choose pada fungsi draw
        draw(choose)
        # memanggil dan memberikan nilai rumah_anayah, "C", "G" pada fungsi draw
        maps = shortest_path(rumah_anayah,"C","G")
        # cetak tulisan dan fungsi maps
        print ("\nJalur Dari 'C' ke 'G' :", maps)
    # jika user memilih 2
    elif (choose == 2):
      # tulisan dibawah akan dicetak
        print ("\nRumah Ari ke UPJ:")
        # memanggil dan memberikan nilai choose pada fungsi draw
        draw(choose)
        # memanggil dan memberikan nilai rumah_ari, "A", "G" pada fungsi draw
        maps = shortest_path(rumah_ari,"A","G")
        # cetak tulisan dan fungsi maps
        print ("\nJalur Dari 'A' ke 'G' :", maps)
    # jika user memilih 3
    elif (choose == 3):
      # tulisan dibawah akan dicetak
        print ("\nRumah Banez ke UPJ:")
        # memanggil dan memberikan nilai choose pada fungsi draw
        draw(choose)
        # memanggil dan memberikan nilai rumah_banez, "K", "O" pada fungsi draw
        maps = shortest_path(rumah_banez,"K","O")
        # cetak tulisan dan fungsi maps
        print ("\nJalur Dari 'K' ke 'O' :", maps)
    # jika user memilih 4
    elif (choose == 4):
      # tulisan dibawah akan dicetak
        print ("\nRumah Bagoes ke UPJ:")
        # memanggil dan memberikan nilai choose pada fungsi draw
        draw(choose)
        # memanggil dan memberikan nilai rumah_bagoes, "A", "I" pada fungsi draw
        maps = shortest_path(rumah_bagoes,"A","I")
        # cetak tulisan dan fungsi maps
        print ("\nJalur Dari 'A' ke 'I' :", maps)
    # jika user memilih 5
    elif (choose == 5):
      # tulisan dibawah akan dicetak
        print ("\nRumah Triana ke UPJ:")
        # memanggil dan memberikan nilai choose pada fungsi draw
        draw(choose)
        # memanggil dan memberikan nilai rumah_triana, "R", "X" pada fungsi draw
        maps = shortest_path(rumah_triana,"R","X")
        # cetak tulisan dan fungsi maps
        print ("\nJalur Dari 'R' ke 'X' :", maps)
    # jika user memilih angka selain 1-5, maka tulisan dibawah akan dicetak
    else :
        print ("Pilihan tidak tersedia")

# jika __name__ = '__main__'
if __name__ == '__main__':
    # maka, panggil fungsi _main()
    _main()
```
