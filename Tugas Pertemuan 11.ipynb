```python
import networkx as nx 
import matplotlib.pyplot as plt

G = nx.DiGraph() 
G.add_edges_from([(1, 2), (2, 3), 
                  (2, 4), (3, 2), 
                  (3, 1), (4, 1), 
                  (1,4)]) 
  
plt.figure(figsize =(5, 5)) 
nx.draw_networkx(G, with_label = True, node_color ='green') 
```
